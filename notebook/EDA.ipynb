```python
df['total_score']=df['math_score']+df['reading_score']+df['writing_score']
```

```python
math_total=0
for a in df['math_score']:
    math_total=math_total+a
print(f"AVERAGE OF MATHS SCORE IS {math_total/1000}")
writing_total=0
for a in df['writing_score']:
    writing_total=writing_total+a
print(f"AVERAGE OF WRITING SCORE IS {writing_total/1000}")
reading_total=0
for a in df['reading_score']:
    reading_total=reading_total+a
print(f"AVERAGE OF READING SCORE IS {reading_total/1000}")
```

```python
fig,axs=plt.subplots(1,2,figsize=(15,7))
sns.histplot(data=df,x='average_score',bins=15,kde=True,color='g',ax=axs[0])
sns.boxplot(data=df,x='average_score',ax=axs[1])
```

```python
fig,axs=plt.subplots(1,2,figsize=(15,7))
sns.histplot(data=df[df['gender']=='female'],x='average_score',bins=15,kde=True,color='g',ax=axs[0])
sns.histplot(data=df[df['gender']=='male'],x='average_score',bins=15,kde=True,color='r',ax=axs[1])
axs[0].set_title("FEMALE")
axs[1].set_title("MALE")
```

```python
fig,axs=plt.subplots(3,2,figsize=(15,7))
sns.histplot(data=df,x='average_score',bins=15,kde=True,color='g',ax=axs[0][0],hue='gender')
sns.histplot(data=df,x='average_score',bins=15,kde=True,color='g',ax=axs[0][1],hue='lunch')
sns.histplot(data=df,x='average_score',bins=15,kde=True,color='g',ax=axs[1][0],hue='parental_level_of_education')
sns.histplot(data=df,x='average_score',bins=15,kde=True,color='g',ax=axs[1][1],hue='test_preparation_course')
sns.histplot(data=df,x='average_score',bins=15,kde=True,color='g',ax=axs[2][0],hue='race_ethnicity')

```

1. Average scores are approximately normally distributed.

2. Most students have average scores between 60 and 80.

3. Female students show a slightly higher concentration around 70–75 marks compared to male students.

4. Very few students have average scores below 30.

5. The performance distribution of both genders is largely similar, though females appear to perform slightly better on average.
6. Average score don not much dependent on parental education . average score is more or less same for everyone
7. also students who have completed their test prepation course have better average score as compared to those who did not
8. Students with standard lunch have better average score as compared to free or reduced lunch
9. Students from group c,d and group e scored more as compared to  a and b
